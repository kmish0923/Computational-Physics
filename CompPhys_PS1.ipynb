```python
#Import Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
plt.style.use("fivethirtyeight")

#Define function to calculate gaussian given a minimum adn maximum x-value, mean, standard deviation

def plotGaussian(minimum, maximum, mu, sigma):
    #create an array to calculate  gaussian function from
    x = np.arange(minimum, maximum, .00001)
    
    #calculate the normalizing coefficient for the Gaussian function
    coeff = 1/(sigma*((2*np.pi)**(.5)))
    
    #Calculate y values of function 
    y = coeff*np.exp(-.5*((x-mu)/sigma)**2)
    
    #plot values and define and plot labels
    plt.plot(x,y)
    plt.title('Normalized Gaussian Curve', fontname = 'Times New Roman')
    plt.xlabel('x [UNITS]', fontname = 'Times New Roman')
    plt.ylabel('y [UNITS]', fontname = 'Times New Roman')
    plt.show()

#Plot a normalized Gaussian Function
plotGaussian(-10, 10, 0, 3)
```
